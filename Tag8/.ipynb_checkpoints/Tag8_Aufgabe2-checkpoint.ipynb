```python
file = open('Aufgabe8.csv', 'r')
Lines = file.readlines()
```

```python
i=0
while i< len(Lines):
    line = str(Lines[i])
    line = line.replace('\n', '')
    line = line.split(' ')
    Lines[i]=line
    i+=1
```

```python
LinesCopy = [x[:] for x in Lines]
```

```python
def durchgehen(Zeilennummer):
    global Zeile
    global ZeileVorher
    global accumulator
    
    if Zeile == 628 and ZeileVorher == 627:
        print('Das Ergebnis ist: ' + str(accumulator))
        return 99999
    befehl = Lines[Zeilennummer][0]
    if Zeile in benutzteZeilen:
        return 88888
    else:
        benutzteZeilen.append(Zeile)
        ZeileVorher = Zeile
        if befehl == 'jmp':
            Zeile += int(Lines[Zeile][1])
        elif befehl == 'nop':
            Zeile += 1
        elif befehl == 'acc':
            accumulator += int(Lines[Zeile][1])
            Zeile += 1
        durchgehen(Zeile)
```

```python
def main():
    global Lines
    global benutzteZeilen
    global Zeile
    global accumulator
        
    
    for i in range(len(Lines)):
        accumulator = 0
        Lines = [x[:] for x in LinesCopy]
        print('i ist: ' + str(i))
        benutzteZeilen = []
        Zeile = 0
        
        if Lines[i][0] == 'jmp':
            Lines[i][0] = 'nop'
        elif Lines[i][0] == 'nop':
            Lines[i][0] = 'jmp'
        
            
        if durchgehen(Zeile) == 99999:
            print('Das Ergebnis ist: ' + str(accumulator))
            return
    
```

```python
main()
```
