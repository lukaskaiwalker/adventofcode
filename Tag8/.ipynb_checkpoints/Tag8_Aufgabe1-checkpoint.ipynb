```python
file = open('Aufgabe8.csv', 'r')
Lines = file.readlines()
```

```python
i=0
while i< len(Lines):
    line = str(Lines[i])
    line = line.replace('\n', '')
    line = line.split(' ')
    Lines[i]=line
    i+=1
Lines
```

```python
accumulator = 0
Zeile = 0
benutzteZeilen = []
```

```python
def durchgehen(befehl):
    global Zeile
    global accumulator
    if Zeile in benutzteZeilen:
        print(accumulator)
        return
    else:
        benutzteZeilen.append(Zeile)
        if befehl == 'jmp':
            Zeile += int(Lines[Zeile][1])
            durchgehen(Lines[Zeile][0])
        elif befehl == 'nop':
            Zeile += 1
            durchgehen(Lines[Zeile][0])
        elif befehl == 'acc':
            accumulator += int(Lines[Zeile][1])
            Zeile += 1
            durchgehen(Lines[Zeile][0])
```
