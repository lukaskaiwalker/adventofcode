```python
file = open('Aufgabe7.csv', 'r')
Lines = file.readlines()

```

```python
i=0
while i< len(Lines):
    line = str(Lines[i])
    line = line.replace(' bags contain ', ', ')
    line = line.split(', ')
    Lines[i]=line
    i+=1
Lines
```

```python
Taschen = ['shiny gold']
```

```python
def SucheTasche(bag_name):
    i = 0
    j = 1
    #gehe durch alle Zeilen:
    while i < len(Lines):
        neueTasche = Lines[i][0]
        j = 1
        #gehe durch alle Inhalte der Zeile i:
        while j < len(Lines[i]):
            #Check, ob die Tasche sich darin befindet:
            if Lines[i][j].find(bag_name) != -1 and neueTasche not in Taschen:
                Taschen.append(neueTasche)
                SucheTasche(neueTasche)
            j=j+1
        i=i+1
            
```

```python
SucheTasche('shiny gold')
```

```python
print(Taschen)
```

```python
print('Die Antwort ist: ' + str(len(Taschen)-1))
#die initiale Tasche darf nicht gezählt werden
```
