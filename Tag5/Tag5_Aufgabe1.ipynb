```python
file= open('Aufgabe5.csv', 'r')
Lines = file.readlines()
Lines
```

```python
max=0
for line in Lines:
    actual=0
    
    row = line[:7]
    rowNum=row.replace('F','0')
    rowNum2=rowNum.replace('B','1')
    row=int(rowNum2, 2)
    
    column= line [7:10]
    colNum=column.replace('L','0')
    colNum2=colNum.replace('R','1')
    column=int(colNum2, 2)
    
    actual = row * 8 + column
    if actual>max:
        max=actual
print(max)
```
