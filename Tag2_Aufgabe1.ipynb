```python
import pandas as pd
```

```python
delimiter_amount = "-"
delimiter_char = ":"
```

```python
df=pd.read_csv('//HOLMFS06/User-3/I054569/python/adventofcode/Tag2/aufgabe2.csv', sep=" ", names=["Amount", "Char", "Password"])
df
```

```python
new=df["Amount"].str.split(delimiter_amount, n = 1, expand = True)
df["Amount_from"]=new[0]
df["Amount_to"]=new[1]
new=df["Char"].str.split(delimiter_char, n=1, expand=True)
df["Char"]=new[0]
df
```

```python
i = 0
count= 0
df["Password_Right"]=0
df["Count"]=0
while i < len(df):
    for c in df["Password"][i]: 
        if c == df["Char"][i]: 
            count = count + 1
    if count >= int(df.loc[i, "Amount_from"]) and count <= int(df.loc[i, "Amount_to"]):
        df.loc[i, "Password_Right"] = 1
    else:
        df.loc[i, "Password_Right"] = 0
    df.loc[i, "Count"]=count
    i += 1
    count=0
    
```
